```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sigmoid function and its derivatives
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

```python
# Generate z values
z = np.linspace(-10, 10, 100)

# Calculate values
y_sigmoid = sigmoid(z)
y_derivative = sigmoid_derivative(z)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(z, y_sigmoid, label='Sigmoid', color='blue')
plt.plot(z, y_derivative, label='Derivative', color='orange', linestyle='--')

plt.title('Sigmoid Function and its Derivative')
plt.xlabel('z')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return np.where(z > 0, 1, 0)

# Generate z values
z = np.linspace(-10, 10, 1000)

# Calculate values
y_relu = relu(z)
y_derivative = relu_derivative(z)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(z, y_relu, label='ReLU', color='blue')
plt.plot(z, y_derivative, label='Derivative', color='orange', linestyle='--')

plt.title('ReLU Function and its Derivative')
plt.xlabel('z')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def tanh(z):
    return np.tanh(z)

def tanh_derivative(z):
    return 1 - np.tanh(z) ** 2
```

```python
# Create the plot
z = np.linspace(-10, 10, 1000)
y_tangent = tanh(z)
y_derivative = tanh_derivative(z)

plt.figure(figsize=(10, 6))
plt.plot(z, y_tangent, label='hyperbolic tangent', color='blue')
plt.plot(z, y_derivative, label='Derivative', color='orange', linestyle='--')

plt.title('hyperbolic tangent function and its Derivative')
plt.xlabel('z')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

```python

z = np.linspace(-5, 5, 100)
y_tangent = tanh(z)
y_relu = relu(z)
plt.figure(figsize=(10, 6))
plt.plot(z, y_tangent, label=' tanH', color='blue')
plt.plot(z, y_relu, label='ReLU', color='orange')

plt.title('hyperbolic tangent function and its Derivative')
plt.xlabel('z')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

```python
z = np.linspace(-5, 5, 100)

# Get the Relu and tanH gradient values
tanh_grad = tanh_derivative(z)
relu_grad = relu_derivative(z)


# Plot the activation functions
plt.figure(figsize=(12, 6))

# Plot Relu and its derivative
plt.subplot(1, 2, 1)
plt.plot(z, relu(z), label='ReLU Activation', color='g')
plt.plot(z, relu_grad, label="ReLU Derivative", color='r', linestyle='--')
plt.title('ReLU Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()


# Plot tanH and its derivative
plt.subplot(1, 2, 2)
plt.plot(z, tanh(z), label='tanH Activation', color='g')
plt.plot(z, tanh_grad, label="tanH Derivative", color='r', linestyle='--')
plt.title('tanH Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

plt.tight_layout()
plt.show()
```
